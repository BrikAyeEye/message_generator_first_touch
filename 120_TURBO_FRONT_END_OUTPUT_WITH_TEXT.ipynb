```python
# 1
import pandas as pd
import random

# Read data from CSV file
data = pd.read_csv('01_01_Competitors_of_Crelate_Contacts_for_Contacting.csv')

# Convert DataFrame to list of dictionaries
data_dict_list = data.to_dict('records')

#2
# Process the data
managers = []
companies = []

for row in data_dict_list:  # Updated from 'data' to 'data_dict_list'
    manager_info = {
        "first_name": row['First Name'],
        "current_position_title": row['Current Position Job Title'],
        "current_position_company": row['Current Position Company'],
        "email": row['Email Addresses (Work)'],
        "linkedin_url": row['Websites (LinkedIn)'],
        "linkedin_profile_url": row['LinkedIn Profile Url'],  # Added this line
        "positions": [row['Position A'], row['Position B']]
    }
    managers.append(manager_info)
    
    company_info = {
        "company_name": row['Current Position Company'],
        "career_site": row['Career Site'],
        "positions": [row['Position A'], row['Position B']]
    }
    companies.append(company_info)

# Now you have two lists: managers and companies, each containing dictionaries with the desired information

#3 to #11

import ipywidgets as widgets
from IPython.display import display, clear_output

#3 Assuming 'companies' and 'managers' lists have been populated as per previous code snippets
company_names = list(set(company['company_name'] for company in companies))  # Get unique company names

#4 Widgets
company_dropdown = widgets.Dropdown(options=company_names, description='Select Company:')
select_company_button = widgets.Button(description='Select Company')
manager_dropdown = widgets.Dropdown(options=[], description='Select Manager:')
generate_message_button = widgets.Button(description='Generate Message')
output = widgets.Output()


#5 Display company selection initially
display(company_dropdown, select_company_button)

def on_select_company_button_clicked(b):
    print("Select company button clicked")  # Debug print
    clear_output()  # Clear previous widgets
    display(company_dropdown, select_company_button)  # Re-display company selection
    # Populate manager dropdown based on selected company
    selected_company = company_dropdown.value
    manager_options = [(f"{manager['first_name']}_{manager['current_position_title']}_{manager['linkedin_url']}", manager) 
                       for manager in managers if manager['current_position_company'] == selected_company]
    manager_dropdown.options = manager_options
    display(manager_dropdown, generate_message_button)  # Display manager selection and generate message button
    
def on_generate_message_button_clicked(b):
    print("Generate message button clicked")  # Debug print
    with output:
        clear_output()  # Clear previous output
        selected_manager = manager_dropdown.value
        if selected_manager:
            # Assuming 'generate_message' and 'get_career_page_link' are your functions to generate the message and get the career page link
            message = generate_message(selected_manager['first_name'], selected_manager['positions'][0], selected_manager['positions'][1])
            career_page_link = get_career_page_link(selected_manager['current_position_company'])
            print(f"Message: {message}\nCareer Page Link: {career_page_link}")

select_company_button.on_click(on_select_company_button_clicked)
generate_message_button.on_click(on_generate_message_button_clicked)

#6
def on_select_company_button_clicked(b):
    with output:
        print("Select company button clicked")  # Debug print
        clear_output()  # Clear previous widgets
        display(company_dropdown, select_company_button)  # Re-display company selection
        # Populate manager dropdown based on selected company
        selected_company = company_dropdown.value
        manager_options = [(f"{manager['first_name']}_{manager['current_position_title']}_{manager['linkedin_url']}", manager) 
                           for manager in managers if manager['current_position_company'] == selected_company]
        manager_dropdown.options = manager_options
        display(manager_dropdown, generate_message_button)  # Display manager selection and generate message button
#7
# In your on_generate_message_button_clicked function or wherever you wish to generate and display the messages

def on_generate_message_button_clicked(b):
    with output:
        clear_output()  # Clear previous output
        selected_manager = manager_dropdown.value
        if selected_manager:
            # Generate the first message
            message1 = generate_message(
                selected_manager['first_name'],
                selected_manager['positions'][0],
                selected_manager['positions'][1],
                selected_manager['current_position_company']  # Passing company_name here
            )
            
            # Generate the second message
            message2 = generate_message(
                selected_manager['first_name'],
                selected_manager['positions'][0],
                selected_manager['positions'][1],
                selected_manager['current_position_company']  # Passing company_name here
            )
            
            # Get the career page link
            career_page_link = get_career_page_link(
                selected_manager['current_position_company']
            )
            
            # Get the LinkedIn URL
            linkedin_url = selected_manager['linkedin_profile_url']
            
 # ... rest of your code ...

        # Print the messages, career page link, and LinkedIn URL
        print(f"Message 1: {message1}\n")  # Added \n at the end for a line break
        print(f"Message 2: {message2}\n")  # Added \n at the end for a line break
        print(f"Career Page Link: {career_page_link}")
        print(f"LinkedIn URL: {linkedin_url}\n")
        print("---" * 20)  # Print a separator line

# ... rest of your code ...


# 8 Linking buttons to event handlers as before
select_company_button.on_click(on_select_company_button_clicked)
generate_message_button.on_click(on_generate_message_button_clicked)

# 9 Display output below widgets as before
display(output)


# 10
# Helper function to get career page link
def get_career_page_link(company_name):
    for company in companies:
        if company['company_name'] == company_name:
            return company['career_site']
    return None  # Return None if no matching company is found


def get_career_page_link(company_name):
    # Assuming companies is a list of dictionaries where each dictionary contains company information
    for company in companies:
        if company['company_name'] == company_name:
            return company['career_site']
    return None  # Return None if no matching company is found


#11 Your generate_message function and other code follows here...
def generate_message(first_name, position_a, position_b, company_name, shuffle=True, i_limit=2):
    compartments = {
        "greetings": [
            f"Hi {first_name}, Hope you're doing well!",
            f"Hi {first_name}, hope all is well.",
            f"Hi {first_name}, thanks for connecting.",
            f"Hi {first_name}, I hope you're well."
        ],
        "introductions": [
            "{company_name} is on fire and I want to put some coal in the fire :)",
            "{company_name} popped up on my radar and it is exciting.",
            "Helping Israeli companies grow in the US is a bit of specialty of mine.",
            "Wow, {company_name} is a great story - I would love to be a part of.",
        ],
        "offers": [
            f"Positions such as {position_a} and {position_b} are exactly my specialty.",
            f"Can I help you with your open positions {position_a} and {position_b}?",
            f"For roles as {position_a} and {position_b} - we offer interim and success-fee recruitment.",
            f"I might have some suitable candidates for positions {position_a} and {position_b}."
        ],

        "references": [
            f"Happy to provide references at Redis.",
        ],

        "call_to_actions": [
            "A quick chat perhaps?",
            "What do you think?",
            "Have a good one",
            "Is this interesting?."
        ],
        "closures": [
            "Best, Udo",
            "KR, Udo",
        ]
    }

    message_pieces = []
    for key, phrases in compartments.items():
        phrase = random.choice(phrases) if shuffle else phrases[0]
        message_pieces.append(phrase.format(name=first_name, position_a=position_a, position_b=position_b, company_name=company_name))

    message = ' '.join(message_pieces)
    i_count = message.lower().count(' i ')  # Count occurrences of ' I '

    # If i_count exceeds the limit, recursively call the function until a suitable message is generated
    if i_count > i_limit:
        return generate_message(first_name, position_a, position_b, shuffle, i_limit)
    else:
        return message

# ... (rest of your code)


#15 Usage
names = ["Shahar", "Gideon", "Roy", "Dan"]
for name in names:
    print(generate_message(name, "Position A", "Position B", "Current Position Company"))
    print("---")
```
